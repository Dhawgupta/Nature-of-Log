### <a id='saddle_point'>Derivate tests </a>
https://en.wikipedia.org/wiki/Derivative_test
This test states that for stationry points if $f^1(c) = f^2(c) = \dots = f^{(n)}(c) = 0$ and $f^{(n+1)} \neq 0$, then $f^{(n+1)}$ has 4 possibilities\\
1. $f^{(n+1)}(c) < 0 $ n is odd ->  maxima\\
2. $f^{(n+1)}(c) > 0 $ n is odd ->  minima\\
3. $f^{(n+1)}(c) < 0 $ n is even -> point of inflection (decreasing) (possibly a saddle point)
4. $f^{(n+1)}(c) > 0 $ n is even -> point of inflection (increasing) (possibly a saddle point)

Carrying the same point into teh case of logs\\
if say we have $f^1(c) = f^2(c) = \dots = f^{(n)}(c) = 0$ and $f^{(n+1)} \neq 0$ \\
Then looking at $g(x)$ we get 
\begin{align}
g'(x) &= \dfrac{f'(x)}{f(x)} = 0,\\
g''(x)&= \dfrac{f''(x)f(x) - f'(x)^2}{f(x)^2} = \dfrac{f''(x)}{f(x)} = 0,\\
g''(x) &= \dfrac{f'''(x)f(x) - f''(x)^2}{f(x)^2} = \dfrac{f'''(x)}{f(x)} = 0,\\
\end{align}
In the end we are left with 

$g^{(n+1)}(x) = \dfrac{f^{(n+1)}(x)}{f(x)}$, where we can assume that $x = c$ from the above case.
and log is only defined on positive valued funcitons , hence, we can see that $g^{(n+1)}$ wil follow the same properties as f counterpart and will share the saddle points
